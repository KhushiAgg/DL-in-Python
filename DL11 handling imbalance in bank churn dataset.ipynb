```python
cols_to_scale = ['CreditScore', 'Age','Tenure', 'Balance', 'NumOfProducts','EstimatedSalary']

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
```

```python
X = df1.drop('Exited', axis='columns')
y = df1['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state = 5)
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix, classification_report

def ANN(X_train, X_test, y_train, y_test): 
    model = keras.Sequential([
        keras.layers.Dense(12, input_shape=(12,), activation='relu'),
        keras.layers.Dense(7, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    opt = keras.optimizers.Adam(learning_rate=0.01)

    model.compile(optimizer=opt,
                 loss='binary_crossentropy',
                 metrics=['accuracy'])

    model.fit(X_train, y_train, epochs=100)
    
    print(model.evaluate(X_test, y_test))
    
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    
    print("Classification Report: \n", classification_report(y_test, y_preds))
    
    return y_preds
```

```python
y_preds = ANN(X_train, X_test,y_train, y_test)
```

## Under sampling 

```python
# Class count
count_class_0, count_class_1 = df1.Exited.value_counts()

# Divide by class
df_class_0 = df1[df1['Exited'] == 0]
df_class_1 = df1[df1['Exited'] == 1]
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under, df_class_1], axis=0)

print('Random under-sampling:')
print(df_test_under.Exited.value_counts())
```

```python
X = df_test_under.drop('Exited',axis='columns')
y = df_test_under['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(X_train, X_test, y_train, y_test)
```

## Over Sampling

```python
df_class_1_over = df_class_1.sample(count_class_0, replace = True)
df_over = pd.concat([df_class_0, df_class_1_over], axis=0)

print('Random over-sampling:')
print(df_over.Exited.value_counts())
```

```python
X = df_over.drop('Exited',axis='columns')
y = df_over['Exited']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_preds = ANN(X_train,X_test, y_train, y_test)
```

## Method 4 Ensemble with undersampling

```python
df1.Exited.value_counts()
```

```python
X = df1.drop('Exited',axis='columns')
y = df1['Exited']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
y_train.value_counts()
```

```python
6370/3
```

```python
df2 = X_train.copy()
df2['Exited'] = y_train
df2.head()
```

```python
df2_class0 = df2[df2.Exited==0]
df2_class1 = df2[df2.Exited==1]
```

```python
def get_train_batch(df_majority, df_minority, start, end):
    df_train = pd.concat([df_majority[start:end], df_minority], axis=0)

    X_train = df_train.drop('Exited', axis='columns')
    y_train = df_train.Exited
    return X_train, y_train    
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1, 0, 2124)

y_pred1 = ANN(X_train, X_test, y_train, y_test)
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1, 2124, 4248)

y_pred2 = ANN(X_train, X_test, y_train, y_test)
```

```python
X_train, y_train = get_train_batch(df2_class0, df2_class1, 4248, 6370)

y_pred3 = ANN(X_train, X_test, y_train, y_test)
```

```python
len(y_pred1)
```

```python
y_pred_final = y_pred1.copy()
for i in range(len(y_pred1)):
    n_ones = y_pred1[i] + y_pred2[i] + y_pred3[i]
    if n_ones>1:
        y_pred_final[i] = 1
    else:
        y_pred_final[i] = 0
```

```python
cl_rep = classification_report(y_test, y_pred_final)
print(cl_rep)
```
